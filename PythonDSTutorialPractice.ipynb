# Matplotlib



```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
x=np.arange(1,50,0.2)
y= np.sin(x)
plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
x=np.arange(1,50,0.2)
y= np.tan(x)
plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
x=np.arange(1,50,0.2)
y= np.cos(x)
plt.plot(x,y,"--",c="r",label='cos-graph')
plt.plot(x,np.sin(x),":",c="b",label='cos-graph')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

# Scatter Plot - Used to represent Non-Continues Data. i.e. Each y is dependent on y, x will have no dependency



```python
x=[1,8,4,3,6,5,2]
y=[2,9,11,5,18,12,4]
plt.scatter(x,y,marker="*",s=150)
```
